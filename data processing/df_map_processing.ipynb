## Notebook pre-processing des données pour la carte

```python
import pandas as pd
import geoviews as gv
import geopandas as gpd #to read the geojson #conda install geopandas
from geoviews import dim

#to create the gorgraphic zones
from shapely.geometry import Polygon
from shapely.ops import cascaded_union

#for map interactions
gv.extension('bokeh') #bokeh as backend
```

#### Data Frame des données utiles

```python
#creation d'un df permettant d'avoir la proportion de temps de parole par année en fonctions des zones de vacances

df = pd.read_csv("proccessed_data.csv")

compteur_zone = {'Year' : [2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]}
df_zone = pd.DataFrame(compteur_zone, columns = ['Year'], index=[2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019])
df_zone['Zone_A_male'] = 0
df_zone['Zone_A_female'] = 0
df_zone['Zone_A_music'] = 0
df_zone['Zone_B_male'] = 0
df_zone['Zone_B_female'] = 0
df_zone['Zone_B_music'] = 0
df_zone['Zone_C_male'] = 0
df_zone['Zone_C_female'] = 0
df_zone['Zone_C_music'] = 0

#on ajoute les colonnes pour les pourcentages, afin de garder les données 'duration' si besoin

df_zone['Zone_A_male_p'] = 0.0
df_zone['Zone_A_female_p'] = 0.0
df_zone['Zone_A_music_p'] = 0.0
df_zone['Zone_B_male_p'] = 0.0
df_zone['Zone_B_female_p'] = 0.0
df_zone['Zone_B_music_p'] = 0.0
df_zone['Zone_C_male_p'] = 0.0
df_zone['Zone_C_female_p'] = 0.0
df_zone['Zone_C_music_p'] = 0.0

#On remplit les colonnes en sommant les poucentages de temps de parole par jour
#on crée une sous base par année qui va nous servir à remplir notre bdd par zone

for i in df_zone['Year'] :
#for i in range(2001,2020) : #idem

    zoneA = (df.loc[df["zone_A"] == True]).loc[(df.loc[df["zone_A"] == True])["year"] == i]
    zoneB = (df.loc[df["zone_B"] == True]).loc[(df.loc[df["zone_B"] == True])["year"] == i]
    zoneC = (df.loc[df["zone_C"] == True]).loc[(df.loc[df["zone_C"] == True])["year"] == i]
    
    df_zone["Zone_A_male"][i] = sum(zoneA["male_duration"])
    df_zone["Zone_A_female"][i] = sum(zoneA["female_duration"])
    df_zone["Zone_A_music"][i] = sum(zoneA["music_duration"])
    df_zone["Zone_B_male"][i] = sum(zoneB["male_duration"])
    df_zone["Zone_B_female"][i] = sum(zoneB["female_duration"])
    df_zone["Zone_B_music"][i] = sum(zoneB["music_duration"])
    df_zone["Zone_C_male"][i] = sum(zoneC["male_duration"])
    df_zone["Zone_C_female"][i] = sum(zoneC["female_duration"])
    df_zone["Zone_C_music"][i] = sum(zoneC["music_duration"])
    
    
for i in df_zone['Year'] :
    A_tot = df_zone['Zone_A_male'][i] + df_zone['Zone_A_female'][i] + df_zone['Zone_A_music'][i]
    B_tot = df_zone['Zone_B_male'][i] + df_zone['Zone_B_female'][i] + df_zone['Zone_B_music'][i]
    C_tot = df_zone['Zone_C_male'][i] + df_zone['Zone_C_female'][i] + df_zone['Zone_C_music'][i]
    
    #on arrondi le poucentage à 4 chiffres après la virgule, afin que le remplissage des régions soit représentatif
    df_zone['Zone_A_male_p'][i] = round(df_zone['Zone_A_male'][i] / A_tot, 2)
    df_zone['Zone_A_female_p'][i] = round(df_zone['Zone_A_female'][i] / B_tot, 2)
    df_zone['Zone_A_music_p'][i] = round(df_zone['Zone_A_music'][i] / A_tot,2)
    
    df_zone['Zone_B_male_p'][i] = round(df_zone['Zone_B_male'][i] / B_tot,2)
    df_zone['Zone_B_female_p'][i] = round(df_zone['Zone_B_female'][i] / B_tot,2)
    df_zone['Zone_B_music_p'][i] = round(df_zone['Zone_B_music'][i] / B_tot,2)
    
    df_zone['Zone_C_male_p'][i] = round(df_zone['Zone_C_male'][i] / C_tot,2)
    df_zone['Zone_C_female_p'][i] = round(df_zone['Zone_C_female'][i] / C_tot,2)
    df_zone['Zone_C_music_p'][i] = round(df_zone['Zone_C_music'][i] / C_tot,2)
```

```python
# creation .csv avec les données pour la carte
df_zone.to_csv("df_zone.csv", index = False)
```

#### Test pour la carte
Création des polygones corresspondant aux zones de vacances françaises : A, B et C.
Puis une façons de tracer la carte avec python.

```python
#Creation des polygones (multipolygon) corresspondant aux zones A, B ou C

#load json avec regions
sf = gpd.read_file('regions.geojson')

polygons_A = [sf["geometry"][2], #27
              sf["geometry"][8], #75
              sf["geometry"][10]] #84

polygons_B = [sf["geometry"][1], #24
              sf["geometry"][3], #28
              sf["geometry"][4], #32
              sf["geometry"][5], #44
              sf["geometry"][6], #52
              sf["geometry"][7], #53
              sf["geometry"][11]] #93

polygons_C = [sf["geometry"][0], #11
              sf["geometry"][9]] #76

u_A = cascaded_union(polygons_A)
u_B = cascaded_union(polygons_B)
u_C = cascaded_union(polygons_C)
```

```python
#df pour la carte

sf_zone = sf[:3]
sf_zone = sf_zone.drop(['code'], axis=1)
sf_zone['nom'] = ["A", "B", "C"]
sf_zone['geometry'] = [u_A, u_B, u_C]
```

```python
# tracé carte avec opts() pour un exemple : 2019

#remplissage du df
sf_zone['value'] = [df_zone["Zone_A_male_p"][2019], df_zone["Zone_B_male_p"][2019], df_zone["Zone_C_male_p"][2019]]

#tracé
deps = gv.Polygons(sf_zone, vdims=['nom','value'])
deps.opts(width=600, height=600, toolbar='above', color=dim('value'),
          colorbar=True, tools=['hover'], aspect='equal')
```
